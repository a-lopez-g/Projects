```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 4000

# Negative-Binomial distributions

# Gen1
r_g1 = 0.49
p_g1 = 0.4
x_g1 = np.random.negative_binomial(r_g1, p_g1, N)

num_var_redund_g1 = 5
r = 0.5 + (1.7 - 0.5) * np.random.rand(num_var_redund_g1)
eps = np.round(np.random.rand(N, num_var_redund_g1) * np.tile(r, (N, 1)))

x_g1_redund = x_g1 + eps
X_g1 = np.column_stack((x_g1, x_g1_redund))
C_X_g1 = np.corrcoef(X_g1, rowvar=False)

names_g1 = ["g1"]
names_g1.extend([f"g1_{i}" for i in range(1, num_var_redund_g1 + 1)])

# Gen2
r_g2 = 1.3
p_g2 = 0.11
x_g2 = np.random.negative_binomial(r_g2, p_g2, N)

num_var_redund_g2 = 10
r = 20 + (25 - 20) * np.random.rand(num_var_redund_g2)
eps = np.round(np.random.rand(N, num_var_redund_g2) * np.tile(r, (N, 1)))

x_g2_redund = x_g2 + eps
X_g2 = np.column_stack((x_g2, x_g2_redund))
C_X_g2 = np.corrcoef(X_g2, rowvar=False)

names_g2 = ["g2"]
names_g2.extend([f"g2_{i}" for i in range(1, num_var_redund_g2 + 1)])

# Gen3
r_g3 = 1.52
p_g3 = 0.06
x_g3 = np.random.negative_binomial(r_g3, p_g3, N)

num_var_redund_g3 = 15
r = 55 + (65 - 55) * np.random.rand(num_var_redund_g3)
eps = np.round(np.random.rand(N, num_var_redund_g3) * np.tile(r, (N, 1)))

x_g3_redund = x_g3 + eps
X_g3 = np.column_stack((x_g3, x_g3_redund))
C_X_g3 = np.corrcoef(X_g3, rowvar=False)

names_g3 = ["g3"]
names_g3.extend([f"g3_{i}" for i in range(1, num_var_redund_g3 + 1)])

# Gen4
r_g4 = 1.23
p_g4 = 0.04
x_g4 = np.random.negative_binomial(r_g4, p_g4, N)

num_var_redund_g4 = 20
r = 100 + (110 - 100) * np.random.rand(num_var_redund_g4)
eps = np.round(np.random.rand(N, num_var_redund_g4) * np.tile(r, (N, 1)))

x_g4_redund = x_g4 + eps
X_g4 = np.column_stack((x_g4, x_g4_redund))
C_X_g4 = np.corrcoef(X_g4, rowvar=False)

names_g4 = ["g4"]
names_g4.extend([f"g4_{i}" for i in range(1, num_var_redund_g4 + 1)])

# Gen5
# Dependiente de Gen1 y Gen2
eps_g1 = np.round(np.random.rand(N) * 2)
eps_g2 = np.round(np.random.rand(N) * 20)
x_g5 = np.round(7 * (x_g1 + eps_g1) + 1 * (x_g2 + eps_g2))
C_x_g5 = np.corrcoef(np.column_stack((x_g5, x_g1, x_g2)), rowvar=False)

names_g5 = ["g5"]

# Gen6
# Dependiente de Gen3 y Gen4
eps_g3 = np.round(np.random.rand(N) * 20)
eps_g4 = np.round(np.random.rand(N) * 20)
x_g6 = np.round(1.3 * (x_g3 + eps_g3) + 1 * (x_g4 + eps_g4))
C_x_g6 = np.corrcoef(np.column_stack((x_g6, x_g3, x_g4)), rowvar=False)

names_g6 = ["g6"]

# Genes independientes
r_gi = 0.49
p_gi = 0.4
x_gi = np.random.negative_binomial(r_gi, p_gi, N)

num_var_redund_gi = 5
r = 2.0 + (2.5 - 2.0) * np.random.rand(num_var_redund_gi)
eps = np.round(np.random.rand(N, num_var_redund_gi) * np.tile(r, (N, 1)))

x_gi_redund = x_gi + eps
X_gi = np.column_stack((x_gi, x_gi_redund))
C_X_gi = np.corrcoef(X_gi, rowvar=False)

names_gi = ["gi"]
names_gi.extend([f"gi_{i}" for i in range(1, num_var_redund_gi + 1)])

# Salida
# Dependiente de Gen3 y Gen5
# Bernoulli distribution

x_o = np.zeros(N)
for i in range(N):
    if x_g3[i] > 25 and x_g5[i] > 30:
        p_o = 0.95
    else:
        p_o = 0.05
    x_o[i] = np.random.binomial(1, p_o)

plt.figure()
plt.scatter(x_g3, x_g5, c=x_o)
plt.show()

names_x_o = ["y"]

X = np.column_stack((X_g1, X_g2, X_g3, X_g4, x_g5, x_g6, X_gi, x_o))
names = np.concatenate((names_g1, names_g2, names_g3, names_g4, names_g5, names_g6, names_gi, names_x_o))

# np.savez('synthetic_data_new.npz', X=X, names=names)
```
