### Boxplots

```python
def box_plot(data, edge_color, fill_color):
    bp = ax.boxplot(data, patch_artist=True)
    
    for element in ['boxes', 'whiskers', 'fliers', 'means', 'medians', 'caps']:
        plt.setp(bp[element], color=edge_color)

    for patch in bp['boxes']:
        patch.set(facecolor=fill_color)       
        
    return bp
```

```python
an_xg_m_score = an_xg_m['tst_score_list']
an_lr_m_score = an_lr_m['tst_score_list']
an_rf_m_core = an_rf_m['tst_score_list']
mi_xg_m_score = mi_xg_m['tst_score_list']
mi_lr_m_score = mi_lr_m['tst_score_list']
mi_rf_m_score = mi_rf_m['tst_score_list']

data = [an_lr_m_score, an_rf_m_core,an_xg_m_score,mi_lr_m_score,mi_rf_m_score,mi_xg_m_score]
fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)
ax.set_xticklabels(['AN LR', 'AN RF',
                    'AN XG', 'MI LR', 'MI RF', 'MI XG'])
plt.rcParams.update({'xtick.labelsize': 20})
plt.rcParams.update({'ytick.labelsize': 20})
ax.set_ylim([0.67, 0.9])
# Creating plot
bp = box_plot(data,'blue','lightblue')
# show plot
plt.show()
```
