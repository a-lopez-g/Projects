# 1. *Células madre mesenquimales derivadas de adiposas*

Tenemos las matrices de conteos: 
* X_0618: conteos de las células sin reprogramar
* X_IPSC_0618: conteos de las células reprogramadas

Las preprocesamos: 
* Calculamos la matriz traspuesta de cada una -> células son filas y genes son columnas
* Añadimos una nueva columna (la última) Y -> será 0 si las células están sin reprogramar y 1 si están reprogramadas

### 1.1 Células madre embrionarias sin reprogramar

```python
# Cargamos los datos
X_0618 = pd.read_csv("./X_no_reprogramadas.txt",header=None)
```

```python
X_0618.head()
```

```python
X_0618.shape
```

```python
# Transponemos la matriz de conteos para que las células sean filas y los genes columnas
X_0618 = X_0618.transpose()
```

```python
X_0618.shape
```

```python
#Añadimos una nueva columna "Y" la cual será la etiqueta y en este caso tendrá el valor 0 (sin reprogramar)
```

```python
X_0618["Y"] = 0
```

```python
X_0618.head()
```
